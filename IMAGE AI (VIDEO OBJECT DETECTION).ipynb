```python
import os
import time
```

```python
execution_path = os.getcwd()
print(execution_path)
```

```python
from imageai.Detection import VideoObjectDetection
```

```python
yolo=VideoObjectDetection()
```

```python
yolo.setModelTypeAsYOLOv3()
```

```python
yolo.setModelPath( os.path.join(execution_path , "yolo.h5"))
```

```python
yolo.loadModel()
```

```python
objects_detected = dict()
objects_detected['name']=list()
objects_detected['probability']=list()
def forFrame(frame_number, output_array, output_count):
  print("FRAME NUMBER->  " , frame_number)
  print("Output for each object : ", output_array)
  for i in range(len(output_array)):
    object_dict = output_array[i]
    print('object_dict', object_dict)

    if object_dict['percentage_probability'] > 0.4:
      print(frame_number,object_dict['name'])
      objects_detected['name'].append(object_dict['name'])
      objects_detected['probability'].append(object_dict['percentage_probability'])

  print("Output count for unique objects : ", output_count)
  print("------------END OF A FRAME --------------")
```

```python
start = time.time()
```

```python
start
```

```python
yolo.detectObjectsFromVideo(input_file_path="/content/VDO-AD-900OblectDetection.mp4",
                                          output_file_path=os.path.join(execution_path, "objects detected"),
                                          frames_per_second=1,per_frame_function=forFrame,
                                          minimum_percentage_probability=40,
                                          log_progress=True)
```

```python
yolo.detectObjectsFromVideo(input_file_path="/content/2-VDO-OblectDetectionDetection.mp4",
                                          output_file_path=os.path.join(execution_path, "objects detected__"),
                                          frames_per_second=10,per_frame_function=forFrame,
                                          minimum_percentage_probability=40,
                                          log_progress=True)
```
